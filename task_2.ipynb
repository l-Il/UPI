**Задание 2**: Составить список учебной группы, включающий Х человек. Для каждого студента указать: фамилию и имя, дату рождения (год, месяц и число), также у каждого студента есть зачетка, в которой указаны: предмет (от трех до пяти), дата экзамена, ФИО преподавателя.
Вывод в виде таблице предмета и даты всех экзаменов

```python
# Импорт библиотеки генерации случайных имён
!pip install faker
from faker import Faker
# Импорт библиотеки для работы со случайными числами
from random import randint, choice, shuffle
# Импорт библиотеки для работы с датами
from datetime import datetime as DT, timedelta
# Импорт библиотеки pandas для таблицы
import pandas as pd
```

```python
# Создаем пустой DataFrame для хранения информации о студентах
columns = ["Фамилия", "Имя", "Дата рождения", "Зачётка"]
df = pd.DataFrame(columns=columns)

# Задаем количество студентов в группе
X = 30

# Функция генерации случайной даты
def get_random_date(start, end):
    delta = end - start
    return start + timedelta(randint(0, delta.days))

# Генерируем список предметов и дат экзаменов
def generate_exams():
  study_list = []
  subjects = ['История', 'Математика', 'Информатика', 'Философия', 'Курсовая']
  shuffle(subjects)
  for _ in range(randint(3,5)):
    exam_name = choice(subjects)
    subjects.remove(exam_name)
    exam_date = get_random_date(DT.strptime('01.06.2024', '%d.%m.%Y'), DT.strptime('30.06.2024', '%d.%m.%Y'))
    teacher = Faker().name()
    study_list.append([exam_name, exam_date, teacher])
  return study_list

study_list = generate_exams()


# Заполняем информацию о каждом студенте
student_data = []
for i in range(X):
    last_name, first_name = Faker().name().split(' ', maxsplit=1)
    birth_date = get_random_date(DT.strptime('01.01.2001', '%d.%m.%Y'), DT.strptime('01.01.2002', '%d.%m.%Y'))
    student_data.append([last_name, first_name, birth_date, study_list])

# Добавляем данные студентов в DataFrame
df = pd.concat([df, pd.DataFrame(student_data, columns=columns)], ignore_index=True)

# Выводим DataFrame с информацией о студентах и их зачетках
print("\nИнформация о студентах и их зачетках:")
df
```

```python
columns = ["Предмет", "Время экзамена", "Преподаватель"]
df = pd.DataFrame(columns=columns)
df = pd.concat([df, pd.DataFrame(study_list, columns=columns)], ignore_index=True)

df
```
